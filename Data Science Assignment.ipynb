# 4. Based on Number of employees classify them into 9 categories and add a column with category number Number of employees: 

###### Category Number 1-10 :1 ; 11-50 : 2 ; 51-100 :3 ; 101-250 :4 ; 251-500 :5 ; 510-1000: 6 ; 1001-5000: 7; 5001-10000:8 ; 10000+: 9

```python
Category_Number=[]
print(type(Table['Employees Count'][0]))
len(Table['Employees Count'])
```

```python
Category_Number=[]
for i in range(60126):
    if Table['Employees Count'][i]>=1 and Table['Employees Count'][i]<=10:
        Category_Number.append(1)
    elif Table['Employees Count'][i]>=11 and Table['Employees Count'][i]<=50:
        Category_Number.append(2)
    elif Table['Employees Count'][i]>=51 and Table['Employees Count'][i]<=100:
        Category_Number.append(3)
    elif Table['Employees Count'][i]>=101 and Table['Employees Count'][i]<=250:
        Category_Number.append(4)
    elif Table['Employees Count'][i]>=251 and Table['Employees Count'][i]<=500:
        Category_Number.append(5)
    elif Table['Employees Count'][i]>=501 and Table['Employees Count'][i]<=1000:
        Category_Number.append(6)
    elif Table['Employees Count'][i]>=1001 and Table['Employees Count'][i]<=5000:
        Category_Number.append(7)
    elif Table['Employees Count'][i]>=5001 and Table['Employees Count'][i]<=10000:
        Category_Number.append(8)
    elif Table['Employees Count'][i]>=10001:
        Category_Number.append(9)
    else:
        Category_Number.append(0)
```
